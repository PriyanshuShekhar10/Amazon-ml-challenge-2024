```python
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt
import cv2
import numpy as np
from PIL import Image
import requests
from io import BytesIO

# Download the image from the URL
image_url = 'https://m.media-amazon.com/images/I/91ovPnc7qyL.jpg'
response = requests.get(image_url)
image = Image.open(BytesIO(response.content))

# Convert image to OpenCV format
image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Enable angle classification and set language to English

# Perform OCR on the image
result = ocr.ocr(image_cv, cls=True)

# Print OCR results
for line in result[0]:
    print(line)

# Draw OCR results on the image
boxes = [line[0] for line in result[0]]
txts = [line[1][0] for line in result[0]]
scores = [line[1][1] for line in result[0]]

# Annotate and display the image with OCR results
annotated_image = draw_ocr(image_cv, boxes, txts, scores)
plt.imshow(annotated_image)
plt.axis('off')
plt.show()
```
