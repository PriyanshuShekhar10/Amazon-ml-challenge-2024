```python
# prompt: load this csv /content/drive/MyDrive/66e31d6ee96cd_student_resource_3/student_resource 3/dataset/test.csv

import pandas as pd

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/66e31d6ee96cd_student_resource_3/student_resource 3/dataset/test.csv')

# Display the first few rows of the DataFrame
df.head()
```

```python
# prompt: now split the csv file according to the column entity_name, and download the csv's

# Group the DataFrame by 'entity_name'
grouped = df.groupby('entity_name')

# Iterate over the groups and save them as separate CSV files
for name, group in grouped:
  filename = f'{name}.csv'
  group.to_csv(filename, index=False)
  # Download the file
  from google.colab import files
  files.download(filename)
```

```python
# prompt: print the shape of each entity csv file

import pandas as pd
from google.colab import files

# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/66e31d6ee96cd_student_resource_3/student_resource 3/dataset/test.csv')

# Display the first few rows of the DataFrame
df.head()


# Group the DataFrame by 'entity_name'
grouped = df.groupby('entity_name')

# Iterate over the groups and save them as separate CSV files
for name, group in grouped:
  filename = f'{name}.csv'
  group.to_csv(filename, index=False)
  # Print the shape of each file
  df_temp = pd.read_csv(filename)
  print(f'{filename}: {df_temp.shape}')
  # Download the file
  files.download(filename)
```

```python
# prompt: split the csv file from /content/depth.csv into batches of 5000 rows

import pandas as pd
from google.colab import files

# Load the CSV file
df = pd.read_csv('/content/depth.csv')

# Calculate the number of batches
batch_size = 5000
num_batches = len(df) // batch_size + (len(df) % batch_size > 0)

# Split the DataFrame into batches
for i in range(num_batches):
  start_idx = i * batch_size
  end_idx = min((i + 1) * batch_size, len(df))
  batch_df = df[start_idx:end_idx]

  # Save the batch as a CSV file
  filename = f'depth_batch_{i+1}.csv'
  batch_df.to_csv(filename, index=False)

  # Download the file
  files.download(filename)
```
